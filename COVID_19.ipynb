# Pre-Processing

we convert the people to 1 whose death_date is given and whose is missing we convert to 2

```python
df['Death']=[2 if i=='9999-99-99' else 1 for i in df.DATE_DIED]
```

```python
df
```

```python
df.Death.value_counts() 
```

```python
df.drop('DATE_DIED',inplace=True,axis=1) 
```

```python
df.PREGNANT.value_counts()
```

```python
df['PREGNANT']=[1 if i== 97 else i for i in df.PREGNANT]
df['PREGNANT']=[1 if i== 98 else i for i in df.PREGNANT]
```

```python
df
```

convert 99,97,98 into null values and fill them with mode if mean of missing values are less than 20%

```python
def missing_treat(x):
     x[(x == 97) | (x==98)| (x==99)] = np.nan
     if x.isna().mean() < 0.2:
        x=x.fillna(x.mode()[0])
     return x   
```

```python
df=df.apply(missing_treat)
df.head(2)
```

```python
df.isna().mean() # NOW 2 columns left whose missing values are more than 20% so we can drop those
```

```python
df.drop(['INTUBED','ICU'],axis=1,inplace=True)
```

```python
df.head(2)
```

Outliers Capping by UPPERLIMIT=99% and lower limit=1%

```python
def outliers_treat(x):
    x=x.clip(lower=x.quantile(0.01),upper=x.quantile(0.99))
    return x

```

```python
df=df.apply(outliers_treat)
```

DRop duplicates rows

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df
```

drop those features which has varaince less than 5%

```python
df.apply(lambda x:x.std()/x.mean())[df.apply(lambda x:x.std()/x.mean())<0.05]
```

Outliers Treatment

Drop those which has zero variance
